```python
import numpy as np
import pandas as pd
#from rdkit.Chem import AllChem as Chem
import seaborn as sns
import matplotlib.pyplot as plt
```

# AiZynthFinder outputs train data

## Group 1 - 2 and 3 steps

```python
data1 = pd.read_csv('data/7Mtrain500-1-default.csv')
data1_par1 = pd.read_csv('data/7Mtrain500-1-par1.csv')
```

```python
true = len(data1[data1.number_of_steps == 2]) + len(data1[data1.number_of_steps == 3]) 
true1 = len(data1_par1[data1_par1.number_of_steps == 2]) + len(data1_par1[data1_par1.number_of_steps == 3]) 
print("Default ", true/(len(data1)/100), end="% \n")
print("Param 1 ", true1/(len(data1_par1)/100), end="% ")
```

```python
fig, ((ax0, ax1)) = plt.subplots(nrows=1, ncols=2, figsize=(10,3))
ax0.hist(data1.number_of_steps)
ax0.set_title('default')
ax1.hist(data1_par1.number_of_steps)
ax1.set_title('new param 1')
ax0.grid()
ax1.grid()
ax0.set_xlim(0, 12)
ax1.set_xlim(0, 12)
fig.tight_layout()
plt.show()
```

## Group 2 - 4, 5 and 6 steps

```python
data2 = pd.read_csv('data/7Mtrain500-2-default.csv')
data2_par1 = pd.read_csv('data/7Mtrain500-2-par1.csv')
```

```python
fig, ((ax0, ax1)) = plt.subplots(nrows=1, ncols=2, figsize=(10,3))
ax0.hist(data2.number_of_steps)
ax0.set_title('default')
ax1.hist(data2_par1.number_of_steps)
ax1.set_title('new param 1')
ax0.grid()
ax1.grid()
ax0.set_xlim(0, 12)
ax1.set_xlim(0, 12)
fig.tight_layout()
plt.show()
```

```python
true = len(data2[data2.number_of_steps < 7]) - len(data2[data2.number_of_steps < 4])
true1 = len(data2_par1[data2_par1.number_of_steps < 7]) - len(data2_par1[data2_par1.number_of_steps < 4])
print("Default ", true/(len(data2)/100), end="% \n")
print("Param 1 ", true1/(len(data2_par1)/100), end="% \n")
#print("Param 2 ", true2/(len(par2)/100), end="% \n")
#print("Param 3 ", true3/(len(par3)/100), end="% \n")
```

## Group 3 - 7 and more steps
